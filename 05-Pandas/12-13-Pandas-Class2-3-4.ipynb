```python
import pandas as pd
cust = pd.read_csv("cust_dimen.csv")
market = pd.read_csv("market_fact.csv")
orders = pd.read_csv("orders_dimen.csv")
prod=pd.read_csv("prod_dimen.csv")
shipping=pd.read_csv("shipping_dimen.csv")
```

```python
print(cust.shape)
print(market.shape)
print(orders.shape)
print(prod.shape)
print(shipping.shape) 
```

```python
print(cust.head())
print("\n")
print(market.head())
print("\n")
print(orders.head())
print("\n")
print(prod.head())
print("\n")
print(shipping.head())
print("\n")
```

```python
market_cust = pd.merge(market,cust,how = 'inner', on ='Cust_id')
```

```python
market_cust.shape
```

```python
market_cust.head()
```

```python
orders.head()
```

```python
market_cust_orders = pd.merge(market_cust,orders,how = 'left', on ='Ord_id')
market_cust_orders.shape
```

```python
market_cust_orders.head()
```

```python
prod.head()
```

```python
market_cust_orders_prod = pd.merge(market_cust_orders,prod,how = 'right', on ='Prod_id')
print(market_cust_orders_prod.shape)
```

```python
market_cust_orders_prod.head()
```

```python
shipping.head()
```

```python
market_cust_orders_prod_ship = pd.merge(market_cust_orders_prod,shipping,how = 'outer', on ='Ship_id')
print(market_cust_orders_prod_ship.shape)
```
